# Edge Impulse Experiment Comparison

This notebook presents the results of three experiments conducted on keyword spotting using different DSP blocks and learning models.
The objective was to compare the accuracy and resource usage of different model configurations on an audio dataset.

### Experiments Conducted:
- **Experiment 1**: Used MFCC (Mel-frequency cepstral coefficients) with a neural network classifier.
- **Experiment 2**: Used MFE (Mel-filterbank energy) with a neural network classifier.
- **Experiment 3**: Used MFE with Transfer Learning (Keyword Spotting) to leverage pre-trained models.
